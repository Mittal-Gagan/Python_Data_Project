Accessing Data

```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#You can use the iloc method on a dataframe to look at certain indexes in your dataframe

df.iloc[0:3]
```

```python
# The loc method is pretty useful for looking at indexes and also names of columns
# in this case I am finding all the columns in between salary_rate and salary_hour_avg 

df.loc[:, 'salary_rate':'salary_hour_avg']
```

```python
# in this code we show all the values using the loc method and also use 
# the dropna method to drop any row that has a salary_rate of NaN
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```
