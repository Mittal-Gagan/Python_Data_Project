```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# The sample method returns a number of random items 
# Every time you run it, a new sample will display 


df.sample(5)
```

```python
# you can specify the random_state key word argument so the same sample will display 

df.sample(5, random_state=42)
```

```python
#The copy method makes a copy of the objects indices and data 
#Because we are doing different calculations on dataframes it is important to make copies of the df

df_altered = df.copy()
```

```python
#Now the objects of the dataframes are different

print(f"{id(df_altered)} is different from {id(df)}")
```
