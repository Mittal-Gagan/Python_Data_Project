Index Management


```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#The three methods that this code will cover is 
#df.reset_index()
#df.set_index()
#df.sort_index()
```

```python
#RESET_INDEX

df_usa = df[df['job_country'] == 'United States']

df_usa.head(6)

#when you run this code you'll notice the index is not very organized 
```

```python
#the reset_index method will reset the index 
#use inplace = True to modify the current dataframe instead of creating a new one

df_usa.reset_index(inplace = True)

df_usa
```

```python
#SET_INDEX 
#sets the dataframe index using an existing column 
```

```python
#you essentially set the index as a column you already have in the dataframe and
#it drops the existing index

df_usa.set_index('index', inplace = True)

df_usa
```
