Pie Charts

```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#plotting the counts of jobs that are work from home 

df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct = '%i%%') 
plt.title('Work from Home Status')
plt.ylabel(' ')
plt.show()
```

```python
#Creating a subplot

fig, ax = plt.subplots(1,3)

#dictionary of each column that I want to make a pie chart for and its title 

dict_columns = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

#for loop that enumerates the items in the dictionary 
#calling the pie method in matplotlib to call each column and its value counts 

for i, (column, title) in enumerate(dict_columns.items()):
    ax[i].pie(df[column].value_counts(), startangle = 90, autopct='%1.1f%%', labels= ['False', 'True'])
    ax[i].set_title(title)
    


```
