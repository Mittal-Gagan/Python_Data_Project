Concatenate Dataframes

```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

df.head(3)
```

```python
# I am going to break the data into 12 different data frames so we can practice the concat method
# to do this I am going to create a dictionary of months

months = df['job_posted_month'].unique()

months
```

```python
#dictionay comprehension to create a dictionary of various data frames seperated my months 

dict_months = {month: df[df['job_posted_month'] == month] for month in months}

dict_months['Jan'].head(3)
```

```python
#using concatenate function to combine data into one data frame

df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)

df_q1.head(3)
```
