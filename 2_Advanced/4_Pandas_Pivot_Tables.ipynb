Pivot Tables

```python
#Here we are going to make a pivot table to of the median salary of the top 3 jobs 
#in the 6 most popular countries in our dataset 
```

```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Using the pivot table method you can find the number of instances of each job_title_short
#using aggfunction=size it will show all rows in a group without worrying about the values

df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
#you can also aggregate the pivot table on a value  

df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
#you can also use the keyword argument columns to add a column to aggregate your data for 

df.pivot_table(values='salary_year_avg', index='job_title_short', columns ='job_country', aggfunc='median')
```

```python
#This table is very  wide and I want to make it long instead
#to do this I can simply change the country to the index and the job_title_short to the columns

df.pivot_table(values='salary_year_avg', index='job_country', columns ='job_title_short', aggfunc='median')
```

```python
#Returns the index of the top 6 values of country names

top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
#creating a new variable to store the original table 
#and then setting that df equal to the dataframe and using the loc method to index 
#the top countries variable
#i create a list of job_titles I want to see and then set the variable 
#equal to same df but only indexing the list I set 

df_job_country_salary = df.pivot_table(
    values = 'salary_year_avg',
    index = 'job_country',
    columns = 'job_title_short',
    aggfunc= 'median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary
```

```python
#Now I want to plot this pivot table 

df_job_country_salary.plot(kind='bar')
plt.xlabel(' ')
plt.ylabel('Median Salary (USD)')  
plt.title('Median Salary By Country and Job Title')  
plt.xticks(rotation=45, ha='right')
plt.show()
```
