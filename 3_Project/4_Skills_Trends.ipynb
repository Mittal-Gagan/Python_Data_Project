### How are in-demand skills trending for Data Analysts? 

#### Methodology 
1) Aggregate skill counts monthly 
2) Re-analyze based on percentage of total jobs 
3) Plot the monthly skill demand 

```python
#Importing Libraries 

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import ast
import seaborn as sns

#Loading Data 

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup 

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#Filtering for data analyst jobs in the United States and creating a copy of the data

df_da_us = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

#Creating a new column in the dataframe that shows the month of the job posting 

df_da_us['job_posted_month_no'] = df_da_us['job_posted_date'].dt.month

#Inspecting data 
df_da_us[['job_title_short', 'job_country', 'job_posted_date', 'job_posted_month_no']].sample(5)
```

```python
df_da_us_exploded = df_da_us.explode('job_skills')
```

```python
#Creating a pivot table of job skills in data analyst job postings by month 
df_da_us_pivot = df_da_us_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_da_us_pivot.loc['Total'] = df_da_us_pivot.sum()

#Sorting the Total row 
df_da_us_pivot = df_da_us_pivot[df_da_us_pivot.loc['Total'].sort_values(ascending=False).index]

#Dropping tht Total row because the sort was successful 
df_da_us_pivot = df_da_us_pivot.drop('Total')

df_da_us_pivot
```

```python
#Now I want to see the percent of each job skill appearing in a month

da_totals = df_da_us.groupby('job_posted_month_no').size()
da_totals
```

```python
#Using the pandas .div method to divide dataframe by a series 

df_da_us_percent = df_da_us_pivot.div(da_totals/100, axis=0)
df_da_us_percent
```

```python
#Changing the numbers of months in my dataframe into their actual names

df_da_us_percent = df_da_us_percent.reset_index()
df_da_us_percent['job_posted_month'] = df_da_us_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_us_percent = df_da_us_percent.set_index('job_posted_month')
df_da_us_percent = df_da_us_percent.drop(columns='job_posted_month_no')

df_da_us_percent
```

```python
df_plot = df_da_us_percent.iloc[:,:5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))


plt.tight_layout()
```
